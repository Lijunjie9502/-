```python
from scipy.optimize import fsolve
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

from pylab import mpl 
mpl.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体 
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
```

# 1.2 作为一种技术的算法

## 练习

### 1.2-2

```python
n = np.arange(1,50,1)
insert_sort = 8 * n ** 2
merge_sort = 64 * n * np.log2(n)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
l1 = ax.plot(n, insert_sort, label='$8{n^2}$')
l2 = ax.plot(n, merge_sort, label='$64n\lg n$')
ax.legend()
```

```python
def f(n):
    return 8 * n ** 2 - 64 * n * np.log2(n)

fsolve(f, 43)
```

- 当 n 小于 43 时，插入排序优于归并排序

### 1.2-3

```python
n = np.arange(1,18,1)
x = 100 * n ** 2
y = 2 ** n
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
l1 = plt.plot(n, x, label='$100n^2$')
l2 = plt.plot(n, y, label='$2^n$')
ax.legend()
```

```python
def f(n):
    return 100 * n ** 2 - 2 ** n

fsolve(f, 14)
```

- n 的最小值为 14 时满足要求
